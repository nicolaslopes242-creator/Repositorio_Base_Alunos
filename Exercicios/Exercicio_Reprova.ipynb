```python
## Pegar a nota da prova teórica
## Pedir pro usuario
## Peidr a nota da prva prática
## Calcule a Média
## Como eu calculo a média ??? Se vc não souber ja reprovou
## Se a média for Maior ou igual a 7, print("Aprovado")
## Se não, REPROVOU
teorica = float(input(" Digite a nota que você tirou na prova teórica"))
pratica = float(input(" Digite a nota que você tirou na prova prática"))

max = teorica + pratica
media = max / 2

if media >= 7:
    print(f"Sua média foi {media} então você foi Aprovado")

else :
    print(f"Sua média foi {media} então você foi REPROVADO")
```
